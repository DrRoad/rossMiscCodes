```python
%matplotlib inline
import pickle
import numpy
import numba
import pandas as pd
from inventoryManagement import *
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns; sns.set(style="ticks", color_codes=True, font_scale=1.1)
sns.set_context("notebook")
```

```python
#Import data
currentInventoryState = pd.read_csv('inputATDataForMC.csv')
vendorStats = pd.read_csv('currentStatsAT.csv')
currentInventoryState = currentInventoryState.loc[currentInventoryState['Unit Cost'] > 0.0001]

#Pick off data for monte-carlo
numParts = len(currentInventoryState)
baseData = np.empty((numParts, 9))
baseData[:,0] = range(numParts)
baseData[:,1] = currentInventoryState['Unit Cost'].values
baseData[:,2] = currentInventoryState['Total Usage']/365.0
baseData[:,3] = (currentInventoryState['Total Usage']/365.0*currentInventoryState['Lead Time (days)']).values
baseData[:,4] = currentInventoryState['Unit Cost']*currentInventoryState['Total Usage']
baseData[:,4] /= baseData[:,4].sum()
baseData[:,5] = currentInventoryState['Total Usage']/currentInventoryState['Total Usage'].sum()
vendorMap = { vendor:i for vendor, i in zip(['A', 'B', 'C', 'D', 'E',], range(5)) }
baseData[:,6] = [ vendorMap[vendor] for vendor in currentInventoryState['Vendor Code'] ]

print "Imported data for %d parts" % numParts
```

### Run Monte-Carlo on A, i, SL

```python
#Index for calculating weighted averages
i_w = 4

@numba.jit
def calculateEOQ(baseData, fixedCosts, holdingCost):
    for i in range(baseData.shape[0]):
        #Calculate the EOQ
        D = baseData[i,2]*365.0
        A = max(0,fixedCosts[int(baseData[i,6])])
        c = baseData[i,1]
        baseData[i,7] = max(1,np.sqrt(2*A*D/(holdingCost*c)))
        
        #Calculate the ordering frequency
        baseData[i,8] = D/baseData[i,7]

@numba.jit
def generateCases(baseData, numTrials, cases):
    numParts = len(baseData)
    cases[:,:,0] = np.random.rand(numParts, numTrials)
    cases[:,:,0] = norm.ppf(cases[:,:,0])

@numba.jit
def evaluateCases(baseData, cases, results, summary, summaryByVendor):
    for i in range(cases.shape[0]):
        #Calculate inventory metrics
        for j in range(cases.shape[1]):
            #Calculate reorder point
            cases[i,j,1] = baseData[i,3] + cases[i,j,0]*np.sqrt(baseData[i,3])
            
            #Calculate other stats
            results[i,j,0] = GPoisson(cases[i,j,1], baseData[i,3])
            summary[j,0] += results[i,j,0]*baseData[i,i_w]
            results[i,j,1] = SPoisson(baseData[i,7], cases[i,j,1], baseData[i,3])
            summary[j,1] += results[i,j,1]*baseData[i,i_w]
            results[i,j,2] = BPoisson(baseData[i,7], cases[i,j,1], baseData[i,3])
            summary[j,2] += results[i,j,2]*baseData[i,i_w]
            results[i,j,3] = IPoisson(baseData[i,7], cases[i,j,1], baseData[i,3], results[i,j,2])
            summary[j,3] += results[i,j,3]*baseData[i,1]
            
            #Calculate stats by vendor
            summaryByVendor[j, 0, baseData[i,6]] += results[i,j,0]*baseData[i,i_w] #Service Level
            summaryByVendor[j, 1, baseData[i,6]] += results[i,j,1]*baseData[i,i_w] #Fill Rate
            summaryByVendor[j, 2, baseData[i,6]] += results[i,j,3]*baseData[i,1] #Inventory Investment

@numba.jit
def runSimulation(baseData, numTrials, fixedCosts, holdingCost):
    numParts = baseData.shape[0]
    cases = np.empty((numParts, numTrials,2))
    calculateEOQ(baseData, fixedCosts, 0.22)
    generateCases(baseData, numTrials, cases)
    results = np.empty((numParts, numTrials, 4))
    summary = np.zeros((numTrials, 4))
    summaryByVendor = np.zeros((numTrials, 3, 5))
    
    evaluateCases(baseData, cases, results, summary, summaryByVendor)
    
    return results, summary, summaryByVendor


def runScenarios(baseData, numTrials, setOfFixedCosts, setOfHoldingCosts):
    scenarios = {}
    scenarioNum = 0
    numScenarios = len(setOfFixedCosts)*len(setOfHoldingCosts)
    for fixedCosts in setOfFixedCostMappings.keys():
        for holdingCost in setOfHoldingCosts:
            print "Running Scenario %d of %d: A = mu + %s*sigma, i_h = %s" % (scenarioNum+1, numScenarios ,fixedCosts, holdingCost)
            results, summary, summByVendor = runSimulation(baseData, numTrials, setOfFixedCostMappings[fixedCosts], holdingCost)
            avgOrderingFreq = 0
            for i in range(baseData.shape[0]):
                avgOrderingFreq += (baseData[i,8]*baseData[i,i_w])
            scenarios[str(fixedCosts)+','+str(holdingCost)] = (results, summary, summByVendor, avgOrderingFreq)
            scenarioNum += 1
    
    print "Complete!"
    return scenarios

def combineResultsIntoDF(scenariosDict):
    keys = scenariosDict.keys()
    dataFrames = []
    dataFramesByVendor = []
    revVendorMap = {i:letter for i, letter in zip(range(5), ['A','B','C','D','E'])}
    for key in keys:
        summary = pd.DataFrame(scenariosDict[key][1], columns=['SL','FR','B','I_Invest'])
        summary['Service Level'] = [ "{:.1f}".format(SL) for SL in summary['SL'] ]
        summary['A'] = key.split(',')[0]
        summary['i_h'] = key.split(',')[1]
        summary['freq'] = scenariosDict[key][3]
        dataFrames.append(summary)
        
        #Compile results by vendor
        vendorDFs = []
        for vendor in range(5):
            vendorSummary = pd.DataFrame(scenariosDict[key][2][:,:,vendor], columns=['SL', 'FR', 'I_Invest'])
            vendorSummary['Service Level'] = [ "{:.1f}".format(SL) for SL in vendorSummary['SL'] ]
            vendorSummary['Vendor ID'] = revVendorMap[vendor]
            vendorSummary['A'] = key.split(',')[0]
            vendorSummary['i_h'] = key.split(',')[1]
            vendorSummary['freq'] = scenariosDict[key][3]
            vendorDFs.append(vendorSummary)
            
        dataFramesByVendor.append(pd.concat(vendorDFs))
        
    completeResults = pd.concat(dataFrames)
    resultsByVendor = pd.concat(dataFramesByVendor)
    
    return completeResults, resultsByVendor
                
```

```python
numTrials = 25000
setOfHoldingRates = [0.10, 0.25, 0.5]
setOfFixedCostMappings = {}
for z in [-0.5,0,0.5,1.5]:
    fixedCosts = {vendorMap[vendor]:(vendorStats[vendor][0]+z*vendorStats[vendor][1]) \
                  for vendor in vendorMap.keys()}
    setOfFixedCostMappings[z] = fixedCosts
```

```python
resultsFromScenarios = runScenarios(baseData, numTrials, setOfFixedCostMappings, setOfHoldingRates)
#pickle.dump(resultsFromScenarios, open('scenarioResults.p', 'wb'))
```

```python
dfOfResults, dfOfResultsByVend = combineResultsIntoDF(resultsFromScenarios)
```
